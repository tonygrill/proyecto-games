La columna item_id tiene un dato nulo, para poder trasformarla int, sustituiremos ese valor nulo por 0

```python
merged_items_reviews = pd.merge(left=df_user_items, right=df_user_reviews, on='user_id', how='inner')
```

La columna item_id de merged_items_reviews es d tipo object, lo cual nos impide hacer un merge los dataframes, la trasformaremos a int

```python
# Transforma la columna 'item_id' a tipo entero
merged_items_reviews['item_id'] = merged_items_reviews['item_id'].astype(int)
```

Uniremos el tercer Dataframes df_steam_games restante con merged_items_reviews, volveremos a utilizar el método inner para mantener la coicidencia entre ambos dataframes.

```python
games_items_reviews = pd.merge(left=merged_items_reviews, right=df_steam_games, on='item_id', how='inner')
```

Como ya hemos mencionado antes en este proyecto, el cliente solicito un despliegue temporal de una API que implementaremos mas adelante en un servicio web gratuito, por razones de memoria y optimización eliminaremos todos los datos nulos presentes en nuestro dataset.

```python
games_items_reviews = games_items_reviews.dropna()
```

Visualizamos información de nuestro Dataframe

```python
games_items_reviews.info()
```

Eliminaremos valores repetidos teniendo el siguiente criterio: un jugador puede jugar varios items o juegos, en este caso eliminaredos duplicados con respesto a user_id e item_id, es decir, vamos a eliminar los repetidos en donde exista mas de un registro con el mismo user_id y el mismo item_id. Sería incongruente mantener mas de un registro con el mismo usuario y el mismo juego.

```python
games_items_reviews = games_items_reviews.drop_duplicates(subset=['user_id', 'item_id']).reset_index(drop=True)
```

```python
games_items_reviews.tail(5)
```

Las columnas item_name y app_name tienen la misma información, por esa razón eliminaremos app_name

```python
games_items_reviews = games_items_reviews.drop('app_name', axis=1)
```

```python
games_items_reviews.info()
```

En el siguiente codigo vamos a tratar a la columna genres y sus valores como si fueras datos dummies, vamos a generar nuevas columnas con las categorias incluidas en la columna genres y asignarles valor binario.

```python
# Usar get_dummies para crear columnas binarias para cada categoría única
df_genres = games_items_reviews['genres'].str.strip("[]").str.replace("'", "").str.get_dummies(sep=', ')

# Concatenar las nuevas columnas al DataFrame original
games_items_reviews = pd.concat([games_items_reviews, df_genres], axis=1)
```

Eliminaremos la columna genres ya que obviamente no las necesitamos

```python
# Eliminar la columna original 'genres' si ya no la necesitas
games_items_reviews = games_items_reviews.drop('genres', axis=1)
```

En la siguinte líne configuramos a pandas para que muestre el dataframe con todas sus columnas

```python
pd.set_option('display.max_columns', None)
```

Las columnas tags y specs solo aportan volumen en nuestro dataset, y de hecho, hay valores que estan presentes en las columnas derivadas de genres, por eso las eliminaremos.

```python
# eliminaremos las columnas tags y specs
games_items_reviews = games_items_reviews.drop('tags', axis=1)
games_items_reviews = games_items_reviews.drop('specs', axis=1)
```
