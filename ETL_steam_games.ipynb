## Aquí realizaremos el ETL al archivo "output_steam_games.json"

Importaremos las librerias necesarias

```python
import json
import pandas as pd
import ast
import numpy as np
```

Procederemos a cargar el archivo

```python
fila = []
with open("output_steam_games.json", encoding= "utf-8") as f:
    for line in f.readlines():
        data = json.loads(line)
        fila.append(data)

# Con la lista ya cargada se procede a crear el dataframe
df_steam_games = pd.DataFrame(fila)
df_steam_games.head()
```

Mostramos información sobre nuestro dataset

```python
df_steam_games.info()
```

Nos pudimos percatar de que las líneas comprendidas entre 0 y 88312 son NaN en todas sus columnas, por esa razón procederemos a eliminarlas

```python
# La siguiente línea reescribe la variable df_steam_games tomando los valores mayores a 88312 y reiniciando
# los indices
df_steam_games = df_steam_games.iloc[88312:].reset_index(drop=True)
df_steam_games.info()
```

```python
df_steam_games
```

A este nivel de nuestro ETL podemos determinar que las columnas 'url', 'reviews_url' se pueden eliminar, así como tambien 'title' debido a que tiene la misma información que 'app_name' y tiene más de 2000 campos nulos adicionales en comparación con 'app_name'

```python
# Eliminar las columnas 'url', 'reviews_url', 'early_access' y 'title' del DataFrame
df_steam_games = df_steam_games.drop(['url', 'reviews_url', 'early_access', 'title'], axis=1)

# Mostrar el DataFrame actualizado
print(df_steam_games.head())
```

Revisando el diccionario de datos proporcionado por el cliente, se pudo observar que el campo 'id' es equivalente al campo 'item_id' de los datasets 'australian_user_reviews.json' y 'australian_users_items.json', para ser coherentes renombraremos la columna 'id' por 'item_id'.

```python
# Cambiar el nombre de la columna
df_steam_games.rename(columns={'id': 'item_id'}, inplace=True)
```

```python
df_steam_games.tail()
```

```python
df_steam_games.info()
```

Para poder transformar la columna price a punto flotante debemos sustituir los NaN por algún valor numérico, en este caso lo haremos por 0

```python
# Rellenamos los valores NaN en la columna 'price' con 0
df_steam_games['price'].fillna(0, inplace=True)
```

```python
df_steam_games.info()

# Podemos ver que la columna price ya no tiene valores NaN
```

Ahora se nos presenta otro inconveniente, y es que tenemos valores alfabéticos en la columna price, lo cual nos impide cambiar el tipo de datos, para este caso vamos a cambiar cualquier valor alfabético prensente por -1, de esta manera podrémos imputar los NaN que en este momento son iguales a 0.

```python
# Reemplaza los valores no numéricos por -1
df_steam_games['price'] = pd.to_numeric(df_steam_games['price'], errors='coerce').fillna(-1)
```

En las siguientes líneas ya seremos capaces de cambiar el tipo de dato de la columna price, calcularemos la media sin incluir el valor 0 en los datos para evitar sesgo e imputaremos los 0 por el valor de la media.

```python
# Convertir la columna 'price' a punto flotante
df_steam_games['price'] = df_steam_games['price'].astype(float)

# Calculamos la media de los valores mayores a 0 en la columna 'price'
mean_price = df_steam_games['price'][df_steam_games['price'] > 0].mean()

# Sustituir los valores 0 por la media
df_steam_games['price'] = df_steam_games['price'].replace(0, mean_price)
```

La mayoría los antiguos valores alfabéticos de la columna price eran free to play, para mantener ese sentido de los juegos gratis ahora su valor será 0.

```python
# Sustituye los valores -1 por 0 en la columna 'price'
df_steam_games['price'] = df_steam_games['price'].replace(-1, 0)
```
